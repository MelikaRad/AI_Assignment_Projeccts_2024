```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
from load_hoda import load_hoda
```

```python
X_train, y_train, X_test, y_test = load_hoda(size = 5)
```

### Q 1.

```python
k_values = range(1, 11)
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)
    accuracies.append(accuracy)

plt.plot(k_values, accuracies, marker='o')
plt.title('Dectection Accuracy vs. k Value')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.grid()
plt.show()
```

<b>Question 2) Finding the Optimal Size</b><br>
Using the best K found in the previous part, determine the optimal image size instead of 5x5. It is recommended not to check all possible numbers initially and to start by searching in larger ranges (e.g., s = 2, 5, 10, 20, 50, 100, 200, 500), then narrow down the search range. We will perform trial and error several times. Then, we will test sizes from 1 to 300 once and declare the optimal size with more confidence.

First trial

```python
best_k = 3

sizes = [2, 5, 10, 20, 50, 100, 200, 500]
accuracies = []

for size in sizes:
    
    X_train, y_train, X_test, y_test = load_hoda(size = size)
    
    model = KNeighborsClassifier(n_neighbors=best_k)
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)
    accuracies.append(accuracy)

plt.figure(figsize=(10, 5))
plt.plot(sizes, accuracies, marker='o')
plt.title('Detection Accuracy vs. Image Size for Hoda Dataset')
plt.xlabel('Image Size')
plt.ylabel('Accuracy')
plt.xscale('log')  
plt.grid()
plt.xticks(sizes)
plt.show()
```

<b>Question 4) Digit detection and classification</b><br>
Download the image below, which contains several numbers on a screen. Then, try to find the area of each number and predict the numbers inside it. Avoid considering areas without numbers through coding. After that, draw rectangles around each predicted number and add text labels to the image using code. The goal is to identify all the numbers in the image. However, having one or two errors in the prediction is acceptable for this exercise.
</div>

```python
img =  cv2.imread('persian-digits.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

noise_removed_img = cv2.GaussianBlur(gray, (15,15), 0)

binary_img = cv2.adaptiveThreshold(noise_removed_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,5,2)
plt.imshow(binary_img, cmap='gray')
```

```python
kernel = np.ones((5, 5), np.uint8)
dilated = cv2.dilate(binary_img, kernel, iterations=1)
plt.imshow(dilated, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
img2 = img.copy()

for c in contours:
    cv2.drawContours(img2, [c], -1, (255,0,0), 3)
    
plt.imshow(img2[:,:,::-1])
```

```python
img4 = img.copy()

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)  
    cv2.rectangle(img4, (x, y), (x + w, y + h), (0, 0, 255), 2)  
    
plt.imshow(img4[:,:,::-1])
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX  
font_scale = 0.7
color = (255, 0, 0)  
thickness = 2  

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt) 
    digit_crop = cv2.resize(dilated[y-5:y+h+5, x-5:x+w+5 ], (14,14)).flatten().reshape(1,-1)
    label = str(model.predict(digit_crop)[0])
    position = (x , y-5) 
    cv2.putText(img4, label, position, font, font_scale, color, thickness)
    
plt.imshow(img4[:,:,::-1])
```
