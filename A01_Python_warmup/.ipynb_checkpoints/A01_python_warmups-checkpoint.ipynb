### 2.

```python
grades_dict = {}

num_of_students = int(input("enter num of students: "))

for _ in range(num_of_students):
    name = input("enter stu name: ")
    score = float(input("enter stu grade: "))
    grades_dict[name] = score

if grades_dict:
    total_score = sum(grades_dict.values())
    average_score = total_score / len(grades_dict)
    print(f"Average of grades: {average_score:.2f}")
else:
    print("no grades were given")
```

### 6.

using time 

```python
import numpy as np
import time

array_length = 500000
random_array = np.random.rand(array_length) * 100  

def find_max_with_for(arr):
    max_value = arr[0]  
    for num in arr:
        if num > max_value:
            max_value = num
    return max_value

start_time_for = time.time()
max_value_for = find_max_with_for(random_array)
end_time_for = time.time()

print(f"max value using for: {max_value_for}")
print(f"time of search using for: {end_time_for - start_time_for} sec")

# اندازه گیری زمان اجرای NumPy
start_time_numpy = time.time()
max_value_numpy = np.max(random_array)
end_time_numpy = time.time()

print(f"max value using NumPy: {max_value_numpy}")
print(f"time of search using NumPy: {end_time_numpy - start_time_numpy} sec")
```

### 10.

```python
import pandas as pd

data = {
    'name': ['newton', 'mary', 'edison', 'angela', 'jessica'],
    'gender': ['M', 'F', 'M', 'F', 'F'],
    'grade': [15, 19, 19, 20, 20]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('students.csv', index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv('students.csv')

mean_score = df['grade'].mean()

max_score = df['grade'].max()

gender_of_max_score = df.loc[df['grade'] == max_score, 'gender'].values[0]

mean_female_score = df[df['gender'] == 'F']['grade'].mean()

# چاپ نتایج
print(f"mean of grades: {mean_score:.2f}")
print(f"max grade: {max_score:.2f}")
print(f"gender of max grade: {gender_of_max_score}")
print(f"mean grades of females: {mean_female_score:.2f}")
```

```python
import matplotlib.pyplot as plt


# mean and std based on gender
gender_stats = df.groupby('gender')['grade'].agg(['mean', 'std']).reset_index()

plt.bar(gender_stats['gender'], gender_stats['mean'], yerr=gender_stats['std'], capsize=5, color=['red', 'pink'])
plt.xlabel('gender')
plt.ylabel('grades mean')
plt.title('grades mean and std based on gender')
plt.xticks(rotation=45)
plt.grid(axis='y')
```
